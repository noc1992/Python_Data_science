## Pandas 의 시각화 기능
판다스의 시리즈나 데이터프레임은 plot 이라는 시각화 메서드를 내장하고 있다. 
plot은 matplotlib를 내부에서 임포트하여 사용한다. 

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                  index=pd.date_range('1/1/2018',periods=100),
                  columns=['A','B','C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title("pandas 의 예")
plt.xlabel('시간')
plt.ylabel('Data')
plt.show()
```

plot 메서드의 kind라는 인수를 바꾸면 여러가지 플롯을 그릴 수 있다. 
- bar
- pie
- hist
- kde
- box
- scatter
- area

```python
iris[:5].plot.bar(rot=0)     # plot.bar 형태로도 사용 가능
plt.title("iris 데이터")
plt.xlabel("data")
plt.ylabel("각 feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0,8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)         # T 트랜스폼 전치연산!!!!!!!
plt.title("각 종별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0,8)
plt.show()
```

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객 수 비율')
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.title('각 feature 값들의 빈도수 hist')
plt.xlabel("데이터 값")
plt.show()
```

```python
iris.plot.kde()
plt.title("각 Feature 값들의 빈도수에 대한 kernal Density Plot")
plt.xlabel("데이터 값")
plt.show()
```

```python
iris.plot.box()
plt.title("각 값들의 빈도수에 대한 박스")
plt.xlabel("feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by= 'species')
plt.tight_layout(pad=10,h_pad=4)
plt.title("각 feature의 종 별 데이터에 대한 박스")
plt.show()
```
